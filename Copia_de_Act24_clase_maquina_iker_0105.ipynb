# El Problema: Gestión Ineficiente de Equipos La falta de un registro estandarizado para la maquinaria industrial genera errores operativos y retrasos en el mantenimiento. Sin un sistema claro, es difícil rastrear las especificaciones técnicas y el estado real de cada equipo.

```python
# Definición de la Clase Maquina
class Maquina:
    def __init__(self, tipo, marca, anoFabricacion, estadoOperativo):
        self.tipo = tipo
        self.marca = marca
        self.anoFabricacion = anoFabricacion
        self.estadoOperativo = estadoOperativo # Ejemplo: "Activo" o "En pausa"

    # Función 1: Encender
    def encender(self):
        self.estadoOperativo = "Encendido"
        print(f"\n[SISTEMA] El equipo {self.tipo} ({self.marca}) se ha ENCENDIDO.")

    # Función 2: Calibrar
    def calibrar(self):
        print(f"[PROCESO] Calibrando sensores de la máquina año {self.anoFabricacion}...")
        print(f"[PROCESO] Calibración de la marca {self.marca} completada con éxito.")

    # Función 3: Realizar Mantenimiento
    def realizarMantenimiento(self):
        print(f"[MANTENIMIENTO] Aplicando protocolo de revisión al tipo: {self.tipo}.")
        print(f"[MANTENIMIENTO] El estado operativo actual es: {self.estadoOperativo}.")

# --- Bloque Interactivo con input() ---

print("--- REGISTRO DE MAQUINARIA INDUSTRIAL ---")

# Datos para la Maquina 1
print("\nDatos de la primera máquina:")
tipo1 = input("Tipo de máquina: ")
marca1 = input("Marca: ")
anio1 = int(input("Año de fabricación: "))
estado1 = input("Estado operativo inicial: ")

# Datos para la Maquina 2
print("\nDatos de la segunda máquina:")
tipo2 = input("Tipo de máquina: ")
marca2 = input("Marca: ")
anio2 = int(input("Año de fabricación: "))
estado2 = input("Estado operativo inicial: ")

# --- Creación de Objetos ---
maquina1 = Maquina(tipo1, marca1, anio1, estado1)
maquina2 = Maquina(tipo2, marca2, anio2, estado2)

# --- Ejecución de Funciones ---
print("\n" + "="*40)
print("EJECUTANDO PROTOCOLOS TÉCNICOS")
print("="*40)

# Acciones Máquina 1
maquina1.encender()
maquina1.calibrar()

# Acciones Máquina 2
maquina2.realizarMantenimiento()
maquina2.encender()
```
